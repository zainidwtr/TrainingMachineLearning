Salah satu aspek inti Matplotlib adalah `matplotlib.pyplot`.

Mari kita mulai dengan mengimpor `Matplotlib` dan `Matplotlib.pyplot` sebagai berikut:

```python
# kami menggunakan backend sebaris
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpl.style.use(['ggplot']) # optional: for ggplot-like style
```

```python
# memfilter kasus Fraud or Forgery untuk dianalisis lebih lanjut dalam assignment ini",
fraud = criminal[criminal["major_category"].isin(["Fraud or Forgery"])]
fraud
```

```python
# pengelompokan kasus fraud berdasarkan borough",
fraud_borough = criminal.groupby('borough')
```

```python
fraud_crime = criminal.groupby(["borough","year"])["value"].sum().reset_index()
fraud_crime
```

```python
# mengubah data tahun untuk menjadi data kolom",
fraud_ = fraud_crime.reset_index().pivot(columns='year', index='borough', values='value')
fraud_
```

```python
# menjumlahkan kasus fraud di setiap tahun
fraud_['Total'] = fraud_.sum(axis=1)
```

```python
# sorting jumlah kasus fraud secara descending
fraud_.sort_values(by='Total', ascending=False, axis=0, inplace=True)
```

```python
# Menampilkan 5 top kriminal fraud
fraud_top5 = fraud_.head(5)
fraud_top5
```

```python
# transpose data frame fraud,
fraud_top5t = fraud_top5.T
fraud_top5t
```

```python
fraud_top = fraud_top5t.drop([2008,2009,2010,'Total'])
fraud_top
```

`Plot kotak` adalah cara untuk merepresentasikan *distribusi* data secara statistik melalui lima dimensi utama:

- **Minimun:** Angka terkecil dalam kumpulan data.
- **Kuartil pertama:** Angka tengah antara `minimum` dan `median`.
- **Kuartil kedua (Median):** Angka tengah dari kumpulan data (diurutkan).
- **Kuartil ketiga:** Angka tengah antara `median` dan `maksimum`.

## Awan Kata <a id="8"></a>


Awan `Word` (juga dikenal sebagai awan teks atau awan tag) bekerja dengan cara yang sederhana: semakin banyak kata tertentu muncul dalam sumber data tekstual (seperti pidato, posting blog, atau basis data), semakin besar dan berani kata tersebut muncul di kata awan.

```python
# install wordcloud
# !conda install -c conda-forge wordcloud --yes

!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
from PIL import Image
    
# save mask to alice_mask
alice_mask = np.array(Image.open('alice_mask.png'))
    
print('Image downloaded and saved!')
```

```python
stopwords = set(STOPWORDS)
```

```python
is_crimes = criminal[(criminal['year']) & (criminal['borough'] == 'City of London')]['major_category'].str.cat(sep=' ')
```

```python
# instantiate a word cloud object
is_crimes_wordcloud = WordCloud(
    background_color='White',
    max_words=4000,
    stopwords=stopwords,
    mask=alice_mask
)

# generate the word cloud
is_crimes_wordcloud.generate(is_crimes)
```

```python
# Write your function below
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# Graded-Funtion Begin (~1 Lines)
plt.imshow(is_crimes_wordcloud, interpolation='bilinear')
# Graded-Funtion End

plt.axis('off')
plt.show()
```

## Folium

Folium adalah pustaka Python yang kuat yang membantu Anda membuat beberapa jenis peta Leaflet. Fakta bahwa hasil Folium bersifat interaktif membuat library ini sangat berguna untuk pembuatan dashboard.

Dari halaman dokumentasi Folium resmi:

> Folium dibangun di atas kekuatan perselisihan data ekosistem Python dan kekuatan pemetaan pustaka Leaflet.js. Manipulasi data Anda dengan Python, lalu visualisasikan di peta Leaflet melalui Folium.

> Folium memudahkan untuk memvisualisasikan data yang telah dimanipulasi dengan Python pada peta Leaflet interaktif. Ini memungkinkan pengikatan data ke peta untuk visualisasi choropleth serta melewatkan visualisasi Vincent/Vega sebagai penanda di peta.

> Pustaka memiliki sejumlah ubin bawaan dari OpenStreetMap, Mapbox, dan Stamen, dan mendukung ubin khusus dengan kunci API Mapbox atau Cloudmade. Folium mendukung overlay GeoJSON dan TopoJSON, serta pengikatan data ke overlay tersebut untuk membuat peta choropleth dengan skema warna color-brewer.

```python
#!conda install -c conda-forge folium=0.5.0 --yes
!pip install folium
import folium

print('Folium installed and imported!')
```

```python
#Titik Lokasi London#
latitude = 51.5287718
longitude = -0.2416804
```

```python
# define the world map
london_map = folium.Map(location=[latitude, longitude], zoom_start=8, tiles='Stamen Terrain')

# display world map
london_map
```

```python
print("Rata-rata Kriminal di Kota-kota london")
print("please zoom in to view the map")
krimi=criminal[['borough','value']]
world_geo = r'https://raw.githubusercontent.com/radoi90/housequest-data/master/london_boroughs.geojson' 
world_map=folium.Map(location=[51.5287718,-0.2416804],zoom_start=10, tiles='Stamen Terrain')

world_map.choropleth(
    geo_data=world_geo,
    name='choropleth',
    data=krimi,
    columns=['borough', 'value'],
    key_on='feature.properties.name',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime rate in boroughs of london'
    
)

#folium.LayerControl().add_to(world_map)
world_map.save('LondonCrime.html')
display(world_map)
```
